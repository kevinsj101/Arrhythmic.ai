```python
from sklearn.model_selection import train_test_split
#importing sklearn modules to split training and testing dataset
```

```python
x = dataset.drop('type', axis = 1).values
y = dataset['type'].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=227)
#splitting training and testing dataset
```

```python
from sklearn import svm
clf = svm.SVC()
clf.fit(x_train, y_train)
svc = svm.SVC()
#fitting training data in the classifier
```

```python
y_pred= clf.predict(x_test)
#Making predictions using classifier
```

```python
train_accuracy = clf.score(x_train, y_train)
test_accuracy = clf.score(x_test, y_test) 
#training and testing accuracy
```

```python
print(train_accuracy)
print(test_accuracy)
```
